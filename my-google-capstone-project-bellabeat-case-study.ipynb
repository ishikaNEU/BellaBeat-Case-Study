## 5. Analyze
### 5.1 Total Distance and Total Steps
<a id="analyzetotal"></a>
First, by checking the relation between distance and steps first, the correlation looks pretty linear, we can conclude total distance almost can represent total steps, 

```r
ggplot(data = daily_activity_3) +
  geom_point(mapping = aes(x = total_distance, y =total_steps))+
  geom_smooth(mapping = aes(x = total_distance, y =total_steps))+
  labs(title="Total distance vs. total steps")
```

### 5.2 Exploring other Relationships in the Dataset through visualization.
<a id="explore"></a>

```r
#Here we also see the correlation between Distance made and calories burnt
ggplot(data = daily_activity_3) +
  geom_point(mapping = aes(x = total_distance, y =calories))+
  geom_smooth(mapping = aes(x = total_distance, y =calories))+
  labs(title="Total distance vs. total calories")


#Third Visualisation
#Here we also see the correlation between Total Steps made and calories burnt
ggplot(data = daily_activity_3) +
  geom_point(mapping = aes(x = total_steps, y =calories))+
  geom_smooth(mapping = aes(x = total_steps, y =calories))+
  labs(title="Total steps vs. total calories")


#Fourth Visualisation
#Let's evaluate the correlation between Total Distance made and Tracker distance
ggplot(data = daily_activity_3) + 
  geom_point(mapping = aes(x = total_distance, y = tracker_distance, fill=id)) + 
   labs(title = "Total Distance and Tracker Distance")
```
